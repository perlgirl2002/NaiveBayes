Grocery Sentiment Prediction

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB
import seaborn as sns

import matplotlib.pyplot as plt
from matplotlib import figure

%matplotlib inline
```

```python
df = pd.read_json('C:\\users\\kim\\downloads\\grocery\\Grocery_and_Gourmet_Food_5.json', lines=True)
```

```python
df.head()
```

```python
df['helpful_really'] = np.where(df.helpful.str[0]>df.helpful.str[-1], 1, 0)
```

```python
df.overall.value_counts()
```

```python
df['label'] = np.where(df.overall >= 3, 1, 0)
```

```python
keywords = ["great", "excellent", "good", "highly", "sweetest", "easy", "yum","delicious","awesome", "love"]
my_list = []
for key in keywords:
    df.reviewText.str.lower()
    comment = df.reviewText.str.lower()
    my_list = np.where(comment.str.contains(str(key), case=False, na=False), '1', '0')
    df[str(key)] = my_list
```

```python
keywords = ["but", "sucks","awful"]
my_list = []
for key in keywords:
    df.reviewText.str.lower()
    comment = df.reviewText.str.lower()
    my_list = np.where(comment.str.contains(str(key), case=False, na=False), '0', '1')
    df[str(key)] = my_list
```

```python
df.head()
```

```python
feature_cols = ['great','excellent','good','highly','sweetest','easy','yum','delicious','awesome','love','but','sucks','awful','helpful_really' ]
features = df[feature_cols]
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, df['label'], test_size=0.4, random_state=0)
```

```python
# Instantiate our model and store it in a new variable.
bnb = BernoulliNB()

# Fit our model to the data.
bnb.fit(X_train, y_train)

# Classify, storing the result in a new variable.
y_pred = bnb.predict(X_test)

# Display our results.
print("Number of mislabeled points out of a total {} points : {}".format(
    df.shape[0],
    (y_test != y_pred).sum()
))
```

```python
bnb.score(X_test, y_test)
```
